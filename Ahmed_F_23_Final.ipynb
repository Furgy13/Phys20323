```python
#2.

import numpy as np 
from numpy import arange 
from pylab import plot, xlabel, ylabel, show 

# np.random.seed(3)

# constants 
NRn = 20000
NPo = 0
NPb214 = 0 
NBi = 0 
NTi = 0 
NPb207 = 0 

tauRn = 3.82 * 24 * 60 * 60
tauPo = 3.1 * 60
tauPb214 = 26.8 * 60
tauBi = 19.9
tauTi = 1.3 * 60 
h = 1.0

pRn = 1 - 2 ** (-h / tauRn)
pPo = 1 - 2 ** (-h / tauPo)
pPb214 = 1 - 2 ** (-h / tauPb214)
pBi = 1 - 2 ** (-h / tauBi)
pTi = 1 - 2 ** (-h / tauTi)
tmax = 3000

# lists of plot points 
tpoints = arange(0.0, tmax, h)
Rnpts = []
Popts = []
Pb214pts = []
Bipts = []
Tipts = []
Pb207pts = []

# Energy variables
energy_Rn = 0
energy_Po = 0
energy_Pb214 = 0
energy_Bi = 0
energy_Ti = 0

# main loop 
for t in tpoints:
    Rnpts.append(NRn)
    Popts.append(NPo)
    Pb214pts.append(NPb214)
    Bipts.append(NBi)
    Tipts.append(NTi)
    Pb207pts.append(NPb207)
    
    decayRn = 0 
    decayPo = 0 
    decayPb214 = 0 
    decayBi = 0 
    decayTi = 0 

    for r in range(NTi):
        if np.random.random() < pTi:
            decayTi += 1 
        NTi -= decayTi 
        NPb207 += decayTi
        energy_Ti += 3

    for s in range(NBi):
        if np.random.random() < pBi:
            decayBi += 1
            if 0 <= np.random.random() <= 0.003:
                NBi -= 1
                NPb207 += 1
                energy_Bi += 9  # Energy from Bi decay (b-type)
            else:
                NBi -= 1
                NTi += 1
                energy_Bi += 4  # Energy from Bi decay (a-type)

    for t in range(NPb214):
        if np.random.random() < pPb214:
            decayPb214 += 1 
        NPb214 -= decayPb214
        NBi += decayPb214
        energy_Pb214 += 8  # Energy from Pb214 decay (R-type)

    for u in range(NPo):
        if np.random.random() < pPo:
            decayPo += 1
            if 0 <= np.random.random() <= 0.0002:
                NPo -= 1
                NPb214 += 1
                energy_Po += 4  # Energy from Po decay (R-type)
            else:
                NPo -= 1
                NBi += 1
                energy_Po += 3  # Energy from Po decay (a-type)

    for v in range(NRn):
        if np.random.random() < pRn:
            decayRn += 1 
        NRn -= decayRn
        NPo += decayRn
        energy_Rn += 9  # Energy from Rn decay (Z-type)

# makes graph 
plot(tpoints, Rnpts, c='m', label='Rn')
plot(tpoints, Popts, c='b', label='Po')
plot(tpoints, Pb214pts, c='g', label='Pb214')
plot(tpoints, Bipts, c='k', label='Bi')
plot(tpoints, Tipts, c='r', label='Ti')
plot(tpoints, Pb207pts, c='y', label='Pb207')
xlabel("Time")
ylabel("Number of atoms")
show()

# Print total energy from each decay type
print("Total Energy from Rn decay (R-type):", energy_Rn, "MeV")
print("Total Energy from Po decay (a-type and b-type):", energy_Po, "MeV")
print("Total Energy from Pb214 decay (Z-type):", energy_Pb214, "MeV")
print("Total Energy from Bi decay (R-type and a-type):", energy_Bi, "MeV")
print("Total Energy from Ti decay (b-type):", energy_Ti, "MeV")
```

```python
#3. 

import numpy as np 
from numpy import arange 
from pylab import plot, xlabel, ylabel, show 

# np.random.seed(3)

# constants 
NRn = 20000
NPo = 0
NPb214 = 0 
NBi = 0 
NTi = 0 
NPb207 = 0 

tauRn = 3.82 * 24 * 60 * 60
tauPo = 3.1 * 60
tauPb214 = 26.8 * 60
tauBi = 19.9
tauTi = 1.3 * 60 
h = 1.0

pRn = 1 - 2 ** (-h / tauRn)
pPo = 1 - 2 ** (-h / tauPo)
pPb214 = 1 - 2 ** (-h / tauPb214)
pBi = 1 - 2 ** (-h / tauBi)
pTi = 1 - 2 ** (-h / tauTi)
tmax = 3000

# lists of plot points 
tpoints = arange(0.0, tmax, h)
Rnpts = []
Popts = []
Pb214pts = []
Bipts = []
Tipts = []
Pb207pts = []

total_energy_list = []
energy_Rn_list = []
energy_Po_list = []
energy_Pb214_list = []
energy_Bi_list = []
energy_Ti_list = []
energy_Pb207_list = []

num_runs = 10

for run in range(num_runs):
    # Reset initial conditions for each run
    NRn = 20000
    NPo = 0
    NPb214 = 0
    NBi = 0
    NTi = 0
    NPb207 = 0

    # Reset energy variables for each run
    energy_Rn = 0
    energy_Po = 0
    energy_Pb214 = 0
    energy_Bi = 0
    energy_Ti = 0
    energy_Pb207 = 0

# main loop 
for t in tpoints:
    Rnpts.append(NRn)
    Popts.append(NPo)
    Pb214pts.append(NPb214)
    Bipts.append(NBi)
    Tipts.append(NTi)
    Pb207pts.append(NPb207)
    
    decayRn = 0 
    decayPo = 0 
    decayPb214 = 0 
    decayBi = 0 
    decayTi = 0 

    for r in range(NTi):
        if np.random.random() < pTi:
            decayTi += 1 
        NTi -= decayTi 
        NPb207 += decayTi
        energy_Ti += 3

    for s in range(NBi):
        if np.random.random() < pBi:
            decayBi += 1
            if 0 <= np.random.random() <= 0.003:
                NBi -= 1
                NPb207 += 1
                energy_Bi += 9  # Energy from Bi decay (b-type)
            else:
                NBi -= 1
                NTi += 1
                energy_Bi += 4  # Energy from Bi decay (a-type)

    for t in range(NPb214):
        if np.random.random() < pPb214:
            decayPb214 += 1 
        NPb214 -= decayPb214
        NBi += decayPb214
        energy_Pb214 += 8  # Energy from Pb214 decay (R-type)

    for u in range(NPo):
        if np.random.random() < pPo:
            decayPo += 1
            if 0 <= np.random.random() <= 0.0002:
                NPo -= 1
                NPb214 += 1
                energy_Po += 4  # Energy from Po decay (R-type)
            else:
                NPo -= 1
                NBi += 1
                energy_Po += 3  # Energy from Po decay (a-type)

    for v in range(NRn):
        if np.random.random() < pRn:
            decayRn += 1 
        NRn -= decayRn
        NPo += decayRn
        energy_Rn += 9  # Energy from Rn decay (Z-type)
        
    # Save total and individual energies for this run
    total_energy_list.append(energy_Rn + energy_Po + energy_Pb214 + energy_Bi + energy_Ti)
    energy_Rn_list.append(energy_Rn)
    energy_Po_list.append(energy_Po)
    energy_Pb214_list.append(energy_Pb214)
    energy_Bi_list.append(energy_Bi)
    energy_Ti_list.append(energy_Ti)
    energy_Pb207_list.append(energy_Pb207)

# Calculate averages and standard deviations
avg_total_energy = np.mean(total_energy_list)
std_total_energy = np.std(total_energy_list)

avg_energy_Rn = np.mean(energy_Rn_list)
std_energy_Rn = np.std(energy_Rn_list)

avg_energy_Po = np.mean(energy_Po_list)
std_energy_Po = np.std(energy_Po_list)

avg_energy_Pb214 = np.mean(energy_Pb214_list)
std_energy_Pb214 = np.std(energy_Pb214_list)

avg_energy_Bi = np.mean(energy_Bi_list)
std_energy_Bi = np.std(energy_Bi_list)

avg_energy_Ti = np.mean(energy_Ti_list)
std_energy_Ti = np.std(energy_Ti_list)

avg_energy_Pb207 = np.mean(energy_Pb207_list)
std_energy_Pb207 = np.std(energy_Pb207_list)

print("Average Total Energy:", avg_total_energy, "MeV")
print("Standard Deviation Total Energy:", std_total_energy, "MeV")

print("Average Energy from Rn decay (Z-type):", avg_energy_Rn, "MeV")
print("Standard Deviation Energy from Rn decay (Z-type):", std_energy_Rn, "MeV")

print("Average Energy from Po decay (a-type and b-type):", avg_energy_Po, "MeV")
print("Standard Deviation Energy from Po decay (a-type and b-type):", std_energy_Po, "MeV")

print("Average Energy from Pb214 decay (Z-type):", avg_energy_Pb214, "MeV")
print("Standard Deviation Energy from Pb214 decay (Z-type):", std_energy_Pb214, "MeV")

print("Average Energy from Bi decay (R-type and a-type):", avg_energy_Bi, "MeV")
print("Standard Deviation Energy from Bi decay (R-type and a-type):", std_energy_Bi, "MeV")

print("Average Energy from Ti decay (b-type):", avg_energy_Ti, "MeV")
print("Standard Deviation Energy from Ti decay (b-type):", std_energy_Ti, "MeV")

# makes graph 
plot(tpoints, Rnpts, c='m', label='Rn')
plot(tpoints, Popts, c='b', label='Po')
plot(tpoints, Pb214pts, c='g', label='Pb214')
plot(tpoints, Bipts, c='k', label='Bi')
plot(tpoints, Tipts, c='r', label='Ti')
plot(tpoints, Pb207pts, c='y', label='Pb207')
xlabel("Time")
ylabel("Number of atoms")
show()

# Print total energy from each decay type
print("Total Energy from Rn decay (R-type):", energy_Rn, "MeV")
print("Total Energy from Po decay (a-type and b-type):", energy_Po, "MeV")
print("Total Energy from Pb214 decay (Z-type):", energy_Pb214, "MeV")
print("Total Energy from Bi decay (R-type and a-type):", energy_Bi, "MeV")
print("Total Energy from Ti decay (b-type):", energy_Ti, "MeV")
```
